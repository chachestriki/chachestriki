```python
#import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns

#load datasets
dailyActivity = pd.read_csv('../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv')
sleepDay = pd.read_csv('../input/fitbit/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv')
weightLogInfo = pd.read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/weightLogInfo_merged.csv")
```

We can see that in weightLogInfo Fat column has 65 out of 67 rows with null values, instead of deleting the entries we are going to delete the entire column as it does not give us anything to work with.

```python
weightLogInfo = weightLogInfo.drop(columns='Fat')
weightLogInfo.head()
```

We’ll also count unique IDs to see if the data contains 30 IDs as the survey claims.

```python
len(pd.unique(dailyActivity["Id"]))
```

As we can see, the initial statement said we had information from 30 IDs but the outputs shows 33. It can either mean that users have more than one device or there are more people being tracked. We will leave the new IDs there as they can mean more precision in the final results.

```python
dailyActivity.info()
```

We can see that there is no total minutes column but columns that sorted the time by activity. We create a total min column

```python
dailyActivity['TotalMinutes'] = dailyActivity['VeryActiveMinutes'] + dailyActivity['FairlyActiveMinutes'] + dailyActivity['LightlyActiveMinutes'] + dailyActivity['SedentaryMinutes']
dailyActivity.head()
```

```python
# converting minutes to hours

dailyActivity["TotalHours"] = round(dailyActivity["TotalMinutes"]/60,2)
dailyActivity.head(10)
```

```python
#Create a new column “day_of_the_week” to represent the day of the week
dailyActivity["ActivityDate"] = pd.to_datetime(dailyActivity["ActivityDate"], format="%m/%d/%Y")
dailyActivity["DayOfTheWeek"] = dailyActivity["ActivityDate"].dt.day_name()
```

```python
dailyActivity["DayOfTheWeek"].head(5)
```

```python
#Rearranging the dataset
cols = ['Id', 'ActivityDate', 'TotalSteps', 'TotalDistance', 'Calories', 'TotalMinutes',
       'TotalHours', 'DayOfTheWeek','TrackerDistance',
       'LoggedActivitiesDistance', 'VeryActiveDistance',
       'ModeratelyActiveDistance', 'LightActiveDistance',
       'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes',
       'LightlyActiveMinutes', 'SedentaryMinutes']
dailyActivity = dailyActivity[cols]
dailyActivity.head()
```

## Sharing visualizations

```python
#Calories Burned for every hour logged time
plt.scatter(dailyActivity.TotalHours,dailyActivity.Calories,alpha = 1.0, c = dailyActivity.Calories,cmap = "Spectral")
plt.title("Number of Calories Burned Vs Number of Hours")
plt.ylabel("Number of calories burned")
plt.xlabel("Number of Hours")
# adding annotations and visuals
mean_calories = 2303.609574
mean_hours = 20.312543
mean_sedentary = 991.210638 / 60
plt.colorbar(orientation = "vertical")
plt.axvline(mean_hours, color = "Blue", label = "Median steps")
plt.axvline(mean_sedentary, color = "Purple", label = "Median sedentary")
plt.axhline(mean_calories, color = "Red", label = "Median hours")
plt.xlabel("Hours logged")
plt.ylabel("Calories burned")
plt.title("Calories burned for every hour logged")
plt.legend()
plt.grid(True)
plt.show()
```

By observing the scatterplot

- Generally, we can say that by looking at the graph, **If logged hours are increasing then the number of calories burnt is also increased. this is mainly due to because of sedentary minutes.**
- We can see the positive (weak) correlation here.
- But there are some outliers at the bottom end of the graph, which can mislead us.


```python
#Number of Steps vs Calories Burned
plt.scatter(dailyActivity.TotalSteps,dailyActivity.Calories, c = dailyActivity.Calories , cmap = "Spectral",alpha =0.9)
mean_steps = 7637.910638
mean_calories = 2303.609574
plt.colorbar(orientation = "vertical")
plt.axvline(mean_steps,color = "orange",label = "Mean Steps")
plt.axvline(mean_calories,color = "green",label = "Mean Calories")
plt.xlabel("Total Number of Steps")
plt.ylabel("Calories Burned")
plt.title("Number of Steps Vs Calories Burned")
plt.grid(True)
plt.legend()
plt.show()
```

- By observing the graph we can say it is positive correlations with some limitations.
- If we see the graph we can clearly identify the outliers that for 0 steps also there is data of calories burnt.
- There is another outlier that a number of steps are > 35,000 but the calories burnt are between 2000–3000. which is a manual data entry error or something.

```python
#Very Active Minutes vs Calories Burned
plt.scatter(dfg.VeryActiveMinutes,dfg.Calories,c=dfg.Calories,cmap="Spectral",alpha=1.0)
plt.colorbar(orientation =  "vertical")
plt.title("Very Active Minutes vs Calories Burned")
plt.xlabel("Very Active Minutes")
plt.ylabel("Calories Burned")
plt.grid(True)
plt.show()
```

- By observing the scatter plot we can clearly say that it is a positive correlation.
- **If Very Active Minutes are increasing the number of calories burned is also increasing.**

```python
#Sedantary Active Minutes vs Calories Burned
plt.scatter(dfg.SedentaryMinutes,dfg.Calories,c=dfg.Calories,cmap="Spectral",alpha=1.0)
plt.colorbar(orientation =  "vertical")
plt.title("SedentaryMinutes vs Calories Burned")
plt.xlabel("SedentaryMinutes")
plt.ylabel("Calories Burned")
plt.grid(True)
plt.show()
```

- By observing the scatter plot we can say that there is no clear correlation, it seems to be linear.
- We can see some outliers again.

```python
# calculating total of individual minutes column
very_active_mins = dailyActivity["VeryActiveMinutes"].sum()
fairly_active_mins = dailyActivity["FairlyActiveMinutes"].sum()
lightly_active_mins = dailyActivity["LightlyActiveMinutes"].sum()
sedentary_mins = dailyActivity["SedentaryMinutes"].sum()
# plotting pie chart
slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ["Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary minutes"]
colours = ["green", "yellow", "blue", "red"]
explode = [0, 0, 0, 0.1]
plt.style.use("default")
plt.pie(slices, labels = labels, 
        colors = colours, wedgeprops = {"edgecolor": "black"}, 
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```

- This pie chart is really interesting, we can see the ratio of sedentary and active position overall.

```python
# calculating total of individual minutes column
very_active_dist = dailyActivity["VeryActiveDistance"].sum()
fairly_active_dist = dailyActivity["ModeratelyActiveDistance"].sum()
lightly_active_dist = dailyActivity["LightActiveDistance"].sum()
sendentary_active_dist = dailyActivity["SedentaryActiveDistance"].sum()

# plotting pie chart
slices = [very_active_dist, fairly_active_dist, lightly_active_dist,sendentary_active_dist]
labels = ["Very active distance", "Fairly active distance", "Lightly active distance", "SedentaryActiveDistance"]
colours = ["green", "skyblue", "orange", 'red']
explode = [0, 0, 0, 0]
plt.style.use("default")
plt.pie(slices, labels = labels, 
        colors = colours, wedgeprops = {"edgecolor": "black"}, 
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Distance")
plt.tight_layout()
plt.show()
```

- We can see that light exercise leads to greater distance done.

```python
# Number of times users logged into the app across the week, plotting histogram:
plt.style.use("default")
plt.figure(figsize=(6,4))
plt.hist(dailyActivity.DayOfTheWeek,bins=7,color ="orange",width = 0.8,edgecolor = "black")
# adding annotations and visuals
plt.xlabel("Day of the week")
plt.ylabel("Frequency")
plt.title("No. of times users logged in app across the week", size= 16)
plt.grid(True)
plt.show()
```

- We can observe that when the weekend is approaching the usage is decreasing.

## Insights:
Here we will give our insights which can be backed by data & visualizations.

1. We have seen that there is a positive correlation between Calories and Very Active Minutes.
2. Users spent more than 690 a day on sedentary positions.
3. The FitBit app is used by the majority of users (81.3 percent) to track sedentary behaviors rather than health practices.
4. People favor tracking their activities on weekdays rather than on weekends, possibly because they spend more time outside during the week and stay indoors on weekends.
5. As there appears to be no linear correlation between Calories and Sedentary position, **Bellabeat app can suggest some short fun activities, workouts and exercises to encourage its users to decrease the time spent in a sedentary position.**
6. It has been observed that on average the users sleep 437.5 minutes per night. Bellabeat may recommend the users to set a target to 8 hours a day (480 minutes).

#### Final Recommendations

Bellabeats have a lot of options to boost the performance of their App:
- Reward users that engage more by doing more exercise.
- Start a newsletter campaign giving tips for living the healthy way.
- Create milestones for users or goals (Reach you goal of 10000 steps).
- Make events on weekends: suggesting acitivites and routes and make competitions between users.
